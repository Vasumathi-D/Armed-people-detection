```python
import cv2

# Load the trained YOLOv8 model
model = YOLO(r"C:\Users\rohin\runs\detect\train4\weights\best.pt")

# Open video file
video_path = r"C:\Users\rohin\chatgpt_yolo\videos\training2.mp4"
cap = cv2.VideoCapture(video_path)

# Get video properties
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(r"C:\Users\rohin\chatgpt_yolo\output_videos\knife_detected.mp4", fourcc, fps, (width, height))

# Define colors for different classes
colors = {
    "gun": (0, 0, 255),      # Red for guns
    "knife": (255, 0, 0)     # Blue for knives
}

# Process video frame-by-frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO inference
    results = model(frame)[0]

    # Draw bounding boxes
    for box in results.boxes:
        class_id = int(box.cls[0])  
        confidence = box.conf[0].item()
        x1, y1, x2, y2 = map(int, box.xyxy[0])  

        # Define class labels from your dataset
        class_names = ["gun", "knife"]
        label = class_names[class_id] if class_id < len(class_names) else "unknown"
        color = colors.get(label, (255, 255, 255))  # Default white if unknown

        # Draw bounding box and label
        cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
        cv2.putText(frame, f"{label} {confidence:.2f}%", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    # Save frame to output video
    out.write(frame)

# Cleanup
cap.release()
out.release()
print("✅ Detection completed! Video saved as 'output_videos/knife_detected.mp4'")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure plots are displayed in Jupyter Notebook
%matplotlib inline

# Load the training log CSV
log_path = r"C:\Users\rohin\runs\detect\train4\results.csv"  # Ensure correct path
df = pd.read_csv(log_path)

# Display available columns to verify
print("Available columns in CSV:", df.columns)

# Define column names (based on your CSV structure)
epoch_col = "epoch"
map50_col = "metrics/mAP50(B)"       # Accuracy metric (mAP@50)
map95_col = "metrics/mAP50-95(B)"    # Accuracy metric (mAP@50-95)

box_loss_col = "train/box_loss"      # Training box loss
cls_loss_col = "train/cls_loss"      # Training classification loss
val_box_loss_col = "val/box_loss"    # Validation box loss
val_cls_loss_col = "val/cls_loss"    # Validation classification loss

### 📌 Plot Accuracy (mAP) vs Epochs ###
plt.figure(figsize=(10, 5))
plt.plot(df[epoch_col], df[map50_col], label="mAP@50", color='blue', marker='o')
plt.plot(df[epoch_col], df[map95_col], label="mAP@50-95", color='orange', marker='s')

plt.xlabel("Epochs")
plt.ylabel("Accuracy (mAP)")
plt.title("Model Accuracy (mAP) vs Epochs")
plt.legend()
plt.grid()
plt.savefig("accuracy_vs_epochs.png")  # Save graph as an image
plt.show()


### 📌 Plot Training & Validation Loss vs Epochs ###
plt.figure(figsize=(10, 5))
plt.plot(df[epoch_col], df[box_loss_col], label="Train Box Loss", color='red', linestyle="-", marker='o')
plt.plot(df[epoch_col], df[cls_loss_col], label="Train Class Loss", color='green', linestyle="-", marker='s')
plt.plot(df[epoch_col], df[val_box_loss_col], label="Val Box Loss", color='purple', linestyle="dashed", marker='^')
plt.plot(df[epoch_col], df[val_cls_loss_col], label="Val Class Loss", color='orange', linestyle="dashed", marker='D')

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training & Validation Loss vs Epochs")
plt.legend()
plt.grid()
plt.savefig("loss_vs_epochs.png")  # Save graph as an image
plt.show()
```
